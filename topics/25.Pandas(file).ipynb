```python
| Access type | Single row, multiple columns               | Multiple rows, multiple columns              |
| ----------- | ------------------------------------------ | -------------------------------------------- |
| `.loc`      | `df.loc[row_index, ['col1','col2']]`       | `df.loc[[row1,row2], ['col1','col2']]`       |
| `.iloc`     | `df.iloc[row_index, [col_pos1, col_pos2]]` | `df.iloc[[row1,row2], [col_pos1, col_pos2]]` |
```

```python
b=a.loc[1:3]
print(b)
```

```python
# Access row with index 2
#a.loc[row_label, column_label] :syntax

c = a.loc[1,"Tamil"]
print(c)
d=a.loc[1,"Tamil"]>=a.loc[2,"Tamil"]
d
```
