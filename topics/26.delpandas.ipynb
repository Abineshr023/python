```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
a["Grade"]=label_encoder.fit_transform(a["Grade"])
a
```

```python
b=a.isnull().sum()
print(b)
```

```python
m=a["Grade"].mode()       #repeated most frequent
m
```

```python
#convert back to alphabet of grade
a["Grade"]=label_encoder.inverse_transform(a["Grade"])
a
```

```python
f=a['Tamil'].mean()
e=a['English'].mean()
m=a['Maths'].mean()
s=a['Science'].mean()
ss=a['Social_Science'].mean()
t=a['Total'].mean()
```

```python
a['Tamil'].fillna(f,inplace = True)
a['English'].fillna(e,inplace=True)
a['Maths'].fillna(m,inplace=True)
a['Science'].fillna(s,inplace=True)
a['Social_Science'].fillna(ss,inplace=True)
a['Total'].fillna(t,inplace=True)
```

```python
c=a.isnull().sum()
print(c)
```

```python
a
```

```python
x=a.drop(['StudentID','Grade','Result','Total'],axis=1)
y=a['Total']                                                      #axis=1 is mention  for full column and 0 for rows
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
#1.model selection 2.import train,test and split
#3.mention or split the data for train and testing
#4.mention the both at one (x_tr,x_ts) to compare with package  and mention value for test 0r train
```

```python
from sklearn.linear_model import LinearRegression

model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```
