```python
#dfs-preorder
class preord:
    def __init__(self,data):
        self.d=data
        self.left=None
        self.right=None
def pre(node):
    if node is None:
        return
    print(node.d,end=",")
    pre(node.left)
    pre(node.right)

r=preord('R')
na=preord('A')
nb=preord('B')
nc=preord('C')
nd=preord('D')
ne=preord('E')
nf=preord('F')
ng=preord('G')

r.left=na
r.right=nb

na.left=nc
na.right=nd

nb.left=ne
nb.right=nf

nc.left=ng
pre(r)
```

```python
#dfs-Inorder
class traver:
    def __init__(self,data):
        self.d=data
        self.left=None
        self.right=None
def inorder(node):
    if node is None:
        return
    inorder(node.left)
    print(node.d,end="->")
    inorder(node.right)
t=traver('R')
na=traver('A')
nb=traver('B')
nc=traver('C')
nd=traver('D')
ne=traver('E')
nf=traver('F')
ng=traver('G')

t.left=na
t.right=nb

na.left=nc
na.right=nd

nb.left=ne
nb.right=nf

nc.left=ng

inorder(t)
```

```python
#postorder
class traver:
    def __init__(self,data):
        self.d=data
        self.left=None
        self.right=None
def post(node):
    if node is None:
        return
    post(node.left)
    post(node.right)
    print(node.d,end="->")
t1=traver('R')
na=traver('A')
nb=traver('B')
nc=traver('C')
nd=traver('D')
ne=traver('E')
nf=traver('F')
ng=traver('G')

t1.left=na
t1.right=nb

na.left=nc
na.right=nd

nb.left=ne
nb.right=nf

nc.left=ng

post(t1)
```
